```python
import pandas as pd

train = pd.read_csv("train_walmart.csv")
features = pd.read_csv("features_walmart.csv")
stores = pd.read_csv("stores_walmart.csv")
```

```python
df = train.merge(features,
                 on=["Store","Date","IsHoliday"],
                 how="left")

df = df.merge(stores,
              on="Store",
              how="left")
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Week"] = df["Date"].dt.isocalendar().week.astype(int)
df["Quarter"] = df["Date"].dt.quarter
```

```python
weekly = df.groupby("Date")["Weekly_Sales"].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
plt.plot(weekly["Date"], weekly["Weekly_Sales"])
plt.title("Weekly Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.grid(True)
plt.show()
```

```python
weekly["Date"] = pd.to_datetime(weekly["Date"])

weekly = weekly.set_index("Date")
```

```python
split = int(len(weekly) * 0.8)

train = weekly.iloc[:split]
test = weekly.iloc[split:]
```

```python
test["Prediction"] = test["Weekly_Sales"].shift(1)
test = test.dropna()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(
    test["Weekly_Sales"],
    test["Prediction"]
)

rmse = np.sqrt(
    mean_squared_error(
        test["Weekly_Sales"],
        test["Prediction"]
    )
)

print(mae)
print(rmse)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train["Weekly_Sales"], order=(5,1,0))

model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps=len(test))
```

```python
mae = mean_absolute_error(
    test["Weekly_Sales"],
    forecast
)

rmse = np.sqrt(
    mean_squared_error(
        test["Weekly_Sales"],
        forecast
    )
)

print("MAE:", mae)
print("RMSE:", rmse)
```
